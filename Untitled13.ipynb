```python
import pandas as pd
```

```python
df1 = pd.read_csv('annex1.csv')
df1.head()
```

```python
df2 = pd.read_csv('annex2.csv')
df2.head()
```

```python
def encode(val):
  if val=='sale':
    return 1
  else:
    return 0
df2.loc[:,'Sale or Return'] = df2.loc[:,'Sale or Return'].apply(encode)
df2.head()
```

```python
def encode(val):
  if val=='No':
    return 0
  else:
    return 1
df2.loc[:,'Discount (Yes/No)'] = df2.loc[:,'Discount (Yes/No)'].apply(encode)
df2.head()
```

```python
df2 = df2.groupby(['Date','Item Code']).agg({'Quantity Sold (kilo)':'sum','Unit Selling Price (RMB/kg)':'sum','Sale or Return':'sum','Discount (Yes/No)':'sum'}).reset_index()
df2 = df2.rename(columns={'Quantity Sold (kilo)':'Total Quantity Sold (kilo)','Unit Selling Price (RMB/kg)':'Total Unit Selling Price (RMB/kg)','Sale or Return':
                  'Total Sales','Discount (Yes/No)':'Total Discount'})
```

```python
df2.head()
```

```python
df3 = pd.read_csv('annex3.csv')
df3.head()
```

```python
df4 = pd.read_csv('annex4.csv')
df4.head()
```

```python
df34 = df3.merge(df4,how='inner',on='Item Code')
df34.head()
```

```python
df12 = df2.merge(df34,how='inner',on=['Item Code','Date'])
df12.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df12['Date'] = pd.to_datetime(df12['Date'])
df = df12.sort_values('Date')
```

```python
df['Year'] = df['Date'].dt.year
yearly_qty = df.groupby('Year')['Total Quantity Sold (kilo)'].sum()
plt.figure(figsize=(10, 5))
plt.plot(yearly_qty.index, yearly_qty.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Quantity Sold (kg)')
plt.title('Total Quantity Sold Per Year')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['Year'] = df['Date'].dt.year
yearly_prices = df.groupby('Year')[[
    'Total Unit Selling Price (RMB/kg)',
    'Wholesale Price (RMB/kg)'
]].mean()
plt.figure(figsize=(10, 5))
plt.plot(yearly_prices.index, yearly_prices['Total Unit Selling Price (RMB/kg)'],
         marker='o', label='Selling Price')
plt.plot(yearly_prices.index, yearly_prices['Wholesale Price (RMB/kg)'],
         marker='o', label='Wholesale Price')
plt.xlabel('Year')
plt.ylabel('Price (RMB/kg)')
plt.title('Average Selling vs Wholesale Price per Year')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
